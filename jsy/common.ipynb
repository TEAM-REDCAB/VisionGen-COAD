```python
import os

# 1. common_patients.txt에서 유효한 환자 ID 세트(Set) 만들기
valid_patients = set()
with open('common_patients.txt', 'r', encoding='utf-8') as f:
    next(f)  # 첫 줄(헤더: patient type) 건너뛰기
    for line in f:
        if line.strip():
            # 탭으로 분리하여 첫 번째 값(patient ID)만 가져오기
            patient_id = line.split('\t')[0].strip()
            valid_patients.add(patient_id)

# 2. gdc_manifest.txt 읽으면서 필터링하여 새 파일로 쓰기
with open('gdc_manifest.txt', 'r', encoding='utf-8') as f_in, \
     open('filtered_manifest.txt', 'w', encoding='utf-8') as f_out:
    
    header = next(f_in)
    f_out.write(header)  # GDC 도구가 인식하도록 원본 헤더 그대로 쓰기
    
    keep_count = 0
    total_count = 0
    
    for line in f_in:
        if line.strip():
            total_count += 1
            columns = line.split('\t')
            filename = columns[1]
            
            # 파일명에서 환자 ID 추출 
            # (예: TCGA-AZ-4313-01Z-00... -> TCGA-AZ-4313)
            patient_id = "-".join(filename.split('-')[:3])
            
            # 유효한 환자 목록에 있는지 확인
            if patient_id in valid_patients:
                f_out.write(line)
                keep_count += 1

print(f"🎉 필터링 완료!")
print(f"총 {total_count}개의 매니페스트 목록 중, 임상 데이터가 있는 {keep_count}개의 파일만 추출되었습니다.")
print(f"이제 생성된 'filtered_manifest.txt' 파일로 GDC 다운로드를 진행하세요.")
```
